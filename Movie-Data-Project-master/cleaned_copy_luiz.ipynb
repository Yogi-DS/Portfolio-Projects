```python
movie_data_5000 = "movie_metadata.csv"
movie_data_df = pd.read_csv(movie_data_5000)
movie_data_df2 = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
#Setting order of columns in a cleaner way and dropping unnecessary columns from movie_data_5000:
new_movie_df = movie_data_df[['movie_title', 'content_rating', 'budget', 'gross', 'imdb_score',
                              'num_critic_for_reviews', 'num_voted_users', 'num_user_for_reviews', 'movie_facebook_likes',
                              'cast_total_facebook_likes', 'actor_1_facebook_likes', 'actor_2_facebook_likes',
                              'actor_3_facebook_likes', 'director_facebook_likes']].copy()
new_movie_df.head()
```

```python
new_movie_df = new_movie_df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
new_movie_df = new_movie_df.rename(columns = {'movie_title':'title'})
```

```python
#Setting order of columns in a cleaner way and dropping unnecessary columns from movie_data_1000:
new_movie_df2 = movie_data_df2[['Title', 'Genre', 'Actors', 'Year', 'Runtime (Minutes)', 'Votes','Metascore']].copy()
new_movie_df2.head()
```

```python
new_movie_df2 = new_movie_df2.rename(columns = {'Title':'title','Genre':'genre','Actors':'actors','Year':'year','Runtime (Minutes)':
                                               'runtime_minutes','Votes':'votes','Metascore':'metascore'})
```

```python
new_movie_df2 = new_movie_df2.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
combined_df = new_movie_df.merge(new_movie_df2, on="title",how="inner").copy()
combined_df.tail(30)
```

```python
#looking for missing values to drop from df
combined_df.count()
```

```python
new_combined_df = combined_df.dropna(how="any")
new_combined_df.count()
```
